```python
import pandas as pd
```

```python
path_to_csv = ['./submissions/submission.csv','./submissions/submission1.csv','./submissions/submission2.csv']
```

```python
df = pd.read_csv(path_to_csv[0])
df1 = pd.read_csv(path_to_csv[1])
df2 = pd.read_csv(path_to_csv[2])
```

### length voting

```python
from collections import Counter

answer = df['Vehicleregistrationplatebymodel'].tolist()
answer1 = df1['Vehicleregistrationplatebymodel'].tolist()
answer2 = df2['Vehicleregistrationplatebymodel'].tolist()

for i in range(len(answer)):
  predictions = [answer[i], answer1[i], answer2[i]]

  # Count the occurrences of each length
  lengths = [len(pred) for pred in predictions]
  votes = Counter(lengths)

  # Get the most common length
  final_length = votes.most_common(1)[0][0]

  print(f'DataTest{i+1}: {final_length}')
```

## label voting

```python
from collections import Counter

answer = df['Vehicleregistrationplatebymodel'].tolist()
answer1 = df1['Vehicleregistrationplatebymodel'].tolist()
answer2 = df2['Vehicleregistrationplatebymodel'].tolist()

final_results = []

for i in range(len(answer)):
    predictions = [answer[i], answer1[i], answer2[i]]

    # Count the occurrences of each length
    lengths = [len(pred) for pred in predictions]
    votes = Counter(lengths)

    # Get the most common length
    final_length = votes.most_common(1)[0][0]

    # Transpose the data to get the predictions for each position
    predictions = list(map(list, zip(*predictions)))

    # Initialize the final prediction
    final_prediction = []

    # Loop through the predictions for each position
    for preds in predictions:
        # Count the votes for each character
        votes = Counter(preds)
        
        # Get the character with the most votes
        final_char = votes.most_common(1)[0][0]
        
        # Add the character to the final prediction
        final_prediction.append(final_char)

    # # Truncate or pad the final prediction to match the final length
    # if len(final_prediction) > final_length:
    #     final_prediction = final_prediction[:final_length]
    # elif len(final_prediction) < final_length:
    #     final_prediction += [''] * (final_length - len(final_prediction))

    final_label = ''.join(final_prediction)

    final_results.append(final_label)

    print(f'DataTest{i+1}: {final_label}, length: {final_length}')
```
